```python
import pandas as pd
import numpy as np
```

Create a pandas dataframe of the csv file:

```python
dataframe = pd.read_csv("conversation_stats.csv")
```

```python
dataframe.head()
```

Create an array for the target column:

```python
target = np.array(dataframe.resolved)
```

```python
print("target:\n{}".format(target))
```

```python
print("Keys of dataframe:\n", dataframe.keys())
```

I decided to drop the conversation id since it is not relevant, the data regarding resolution (because that is my target), and Explain Problem Context, because I will test the regression again to see if it changes if I add it back.

```python
drop_columns = ["conversation_id", "resolved", "resolution_type", "resolution_confidence", "move_Explain-Problem-Context"]
features = dataframe.drop(columns=drop_columns)
```

```python
print("Keys of features:\n", features.keys())
```

When I first ran the regression, I got an error message because a lot of the feature columns included "NaN". So, I replaced them all with 0.

```python
features = features.fillna(0)
```

I created a sparse matrix to double check that I don't have any NaN values left.

```python
from scipy import sparse

sparse_matrix = sparse.csr_matrix(features)
print("\nSciPy sparse CSR matrix:\n", sparse_matrix)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    features, target, stratify=target, random_state=42)
logreg = LogisticRegression().fit(X_train, y_train)
print("Training set score: {:.3f}".format(logreg.score(X_train, y_train)))
print("Test set score: {:.3f}".format(logreg.score(X_test, y_test)))
```

The textbook didn't really explain what to do when I get this error message, so I asked CHATGPT and it said I should increase the max iterations:

```python
logreg100 = LogisticRegression(max_iter=5000).fit(X_train, y_train)
print("Training set score: {:.3f}".format(logreg100.score(X_train, y_train)))
print("Test set score: {:.3f}".format(logreg100.score(X_test, y_test)))
```

That's slightly better. I'm also curious to see if increasing C will make a difference. Maybe the current model is underfitting.

```python
logreg100 = LogisticRegression(C=50, max_iter=5000).fit(X_train, y_train)
print("Training set score: {:.3f}".format(logreg100.score(X_train, y_train)))
print("Test set score: {:.3f}".format(logreg100.score(X_test, y_test)))
```

Maybe decreasing C would help then?

```python
logreg100 = LogisticRegression(C=.001, max_iter=5000).fit(X_train, y_train)
print("Training set score: {:.3f}".format(logreg100.score(X_train, y_train)))
print("Test set score: {:.3f}".format(logreg100.score(X_test, y_test)))
```

Okay, so I'll just keep C at its default then.

Now I will try to run the regression, but adding in the Explain-Problem-Context feature.

```python
features2 = features.copy()
features2["move_Explain-Problem-Context"] = dataframe["move_Explain-Problem-Context"]
```

```python
print("Keys of features2:\n", features2.keys())
```

I have to replace NaN again:

```python
features2 = features2.fillna(0)
```

Now I'll run the regression:

```python
X_train, X_test, y_train, y_test = train_test_split(
    features2, target, stratify=target, random_state=42)
logreg = LogisticRegression(max_iter = 5000).fit(X_train, y_train)
print("Training set score: {:.3f}".format(logreg.score(X_train, y_train)))
print("Test set score: {:.3f}".format(logreg.score(X_test, y_test)))
```

Okay, so the training set score is higher. But the test set score is the same. So maybe the model is too complex?
